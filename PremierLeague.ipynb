DATA ANALYSIS

```python
df_2020 = df[df["season"] == 2020]
df_2021 = df[df["season"] == 2021]
df_2022 = df[df["season"] == 2022]
df_2023 = df[df["season"] == 2023]
df_2024 = df[df["season"] == 2024]
```

2020 SEASON BREAKDOWN

```python
df_2020leaderboard = df_2020.groupby("team").agg(
    goals_scored = ('gf','sum'),
    goals_conceded = ('ga', 'sum'),
    average_xG = ('xg','mean'),
    wins = ('result', lambda x: (x == "W").sum()),
    draws = ('result', lambda x: (x == "D").sum()),
    losses = ('result', lambda x: (x == "L").sum()),
)

# Calculate sum of points the team accumulated over the season
df_2020leaderboard["points"] = (df_2020leaderboard['wins'] * 3) + df_2020leaderboard['draws']

# Calculate the goal difference (Goal scored - Goal conceded)
df_2020leaderboard["goal_difference"] = df_2020leaderboard["goals_scored"] - df_2020leaderboard["goals_conceded"]

# Sort the table in DESC order. Value will be sorted from Highest point and goal difference
df_2020leaderboard = df_2020leaderboard.sort_values(by=['points','goal_difference'], ascending = False)

# Assign the teams with the position they ended on at the end of the season
df_2020leaderboard = df_2020leaderboard.reset_index()
df_2020leaderboard['rank'] = df_2020leaderboard.index + 1

df_2020leaderboard
```

```python
# Scatter plot of GF/xG ratio for each team
avg_goals_per_game = df_2020leaderboard['goals_scored'] / 38
avg_xG_per_game = df_2020leaderboard['average_xG']

# Create a temporary DataFrame for plotting
temp_df = pd.DataFrame({
    'Team': df_2020leaderboard['team'],
    'Goals per Game (GF)': avg_goals_per_game,
    'Expected Goals per Game (xG)': avg_xG_per_game
})

# Convert 'Team' column to string if it's not already
temp_df['Team'] = temp_df['Team'].astype(str)

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Create a bar plot for goals per game
bars1 = ax.bar(temp_df['Team'], temp_df['Goals per Game (GF)'], color='red', width=0.5, label='Goals per Game (GF)', align='edge')

# Create a second y-axis for expected goals
ax2 = ax.twinx()
bars2 = ax2.bar(temp_df['Team'], temp_df['Expected Goals per Game (xG)'], color='blue', width=0.5, label='Expected Goals per Game (xG)', align='center')

# Set labels
ax.set_ylabel("Average Goals per Game (GF)", fontsize=14, color='red')
ax2.set_ylabel("Average Expected Goals per Game (xG)", fontsize=14, color='blue')

# Set title and x-label
ax.set_title('Average Goals and Expected Goals per Game by Team', fontsize=16)
ax.set_xlabel('Teams', fontsize=14)

# Set x-ticks
ax.set_xticks(temp_df['Team'])
ax.set_xticklabels(temp_df['Team'], rotation=90)

# Add legends
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```
